

**What is Binary Search?**

Binary Search is an efficient searching algorithm that works on a sorted array/list by repeatedly dividing the search interval in half.

**How to work?**

Start with the middle element.

If it’s the target → return it.

If the target < middle → search left half.

If the target > middle → search right half.

Repeat until element is found or search space is empty.

```python
#Binary Search Algorithm (Iterative)
def binary_search(arr, key):
    low, high = 0, len(arr) - 1

    while low <= high:
        mid = (low + high) // 2   # find middle index

        if arr[mid] == key:
            return mid  # key found

        elif arr[mid] < key:
            low = mid + 1  # search right half
        else:
            high = mid - 1 # search left half

    return -1  # not found


# Example
arr = [10, 20, 30, 40, 50, 60]
key = 30
result = binary_search(arr, key)

print("Element found at index:", result if result != -1 else "Not found")
```

```python
#Recursive Binary Search in Python
def binary_search_recursive(arr, low, high, key):
    if low > high:
        return -1   # not found

    mid = (low + high) // 2

    if arr[mid] == key:
        return mid
    elif arr[mid] < key:
        return binary_search_recursive(arr, mid + 1, high, key)
    else:
        return binary_search_recursive(arr, low, mid - 1, key)


# Example
arr = [10, 20, 30, 40, 50, 60]
key = 50
result = binary_search_recursive(arr, 0, len(arr) - 1, key)

print("Element found at index:", result if result != -1 else "Not found")
```
